```julia
using Printf
using Core

function P(k)
    if k == 2
        return BigFloat(2, 128)
end
return power(2, k-1) * s(k)
end
```

```julia
function s(k)
    if k == 2
        return BigFloat(1, 128)
    end
    return sqrt(1/2*(1-c(k-1)))
end
```

```julia
function c(k)
    if k==2
        return BigFloat(0, 128)
    end
    return sqrt(1/2*(1+c(k-1)))
end
```

```julia
function power(a, b)
    if b == 0
        return BigFloat(1, 128)
    end
    return a*power(a, b-1)
end
```

```julia
k = 2
while k<=256
    @printf("%llf %d \n", P(k), k)
    k = k+1
end
```
