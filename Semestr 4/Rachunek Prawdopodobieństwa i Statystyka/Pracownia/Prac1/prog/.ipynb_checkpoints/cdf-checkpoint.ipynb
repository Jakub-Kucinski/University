```julia
function trapezoidal_rule(f::Function, a::Float64, b::Float64, n::Int64)
    results = zeros(Float64, n+1)
    h = (b-a)
    exp = 1
    x_i = 0.0
    results[1] = 0.5*h*(f(a) + f(b))
    for k = 1 : n
        sum = 0.0
        h = h/2
        for i = 1 : exp
            sum = sum + f(a+(2*i-1)*h)
        end
        results[k+1] = 0.5 * results[k] + h * sum
        exp = exp*2
    end
    return results
end
```

```julia
function rombergs_method(f::Function, a::Float64, b::Float64, n::Int64)
    R = trapezoidal_rule(f, a, b, n)
    exp4 = 1.0
    for k = 1 : n
        exp4 = exp4 * 4.0
        for i = n :-1: k
            R[i+1] = (exp4 * R[i+1] - R[i])/(exp4 - 1)
        end
    end
    return R[n+1]
end
```

```julia
function cdf(f::Function, romberg_depth::Int64, start_t::Float64, interval_t::Float64, end_t::Float64, constant::Float64)
    ts = zeros(Float64, Int64(((end_t - start_t)/interval_t))+1)
    results = zeros(Float64, Int64(((end_t - start_t)/interval_t))+1)
    i = 1
    for t = start_t : interval_t : end_t
        ts[i] = t
        results[i] = 0.5 + sign(t) * rombergs_method(f, 0.0, abs(t), romberg_depth) * constant
        i = i + 1
    end
    return ts, results
end
```

```julia
using Plots
```

```julia
function plot_cdf(f::Function, romberg_depth::Int64, start_t::Float64, interval_t::Float64, end_t::Float64, constant::Float64)
    ts, results = cdf(f, romberg_depth, start_t, interval_t, end_t, constant)
    pl = plot(ts, results, seriestype = :line, xticks=start_t:0.5:end_t, yticks=0:0.1:1,
        lw=3, legend= :none, title="CDF")
    return pl
end
```
