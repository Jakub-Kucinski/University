```python
def roulette_selection(current_population, current_sigmas, objective_values, population_size, children_size):
# ignore x where objective value is infinity
    mask = objective_values != np.inf
    objective_values = objective_values[mask]
    current_population = current_population[mask]
    current_sigmas = current_sigmas[mask]
    population_size = current_population.shape[0]
    
    fitness_values = objective_values.max() - objective_values
    
    if fitness_values.sum() > 0:
        fitness_values = fitness_values / fitness_values.sum()
    else:
        fitness_values = np.ones(population_size) / population_size
    
    parent_indices = np.random.choice(population_size, children_size, True, fitness_values).astype(np.int64)
    
    return current_population[parent_indices, :].copy(), current_sigmas[parent_indices, :].copy()
```

```python
def mi_plus_lambda(current_population, children_population, current_sigmas, children_sigmas,
                              objective_values, children_objective_values, population_size):
    objective_values = np.hstack([objective_values, children_objective_values])
    current_population = np.vstack([current_population, children_population])
    current_sigmas = np.vstack([current_sigmas, children_sigmas])

    I = np.argsort(objective_values)
    current_population = current_population[I[:population_size], :]
    current_sigmas = current_sigmas[I[:population_size], :]
    objective_values = objective_values[I[:population_size]]
    
    return current_population, current_sigmas, objective_values
```

```python
def lambda_replacement(current_population, children_population, current_sigmas, children_sigmas,
                              objective_values, children_objective_values, population_size):
    I = np.argsort(children_objective_values)
    current_population = children_population[I[:population_size], :]
    current_sigmas = children_sigmas[I[:population_size], :]
    objective_values = children_objective_values[I[:population_size]]
    
    return current_population, current_sigmas, objective_values
```

```python
def ES(conditions, objective_function, d=10, population_size=400, children_size=400, K=0.5, 
       mins=None, maxs=None, number_of_iterations=100, parent_selection=roulette_selection, 
       replacement=mi_plus_lambda, sigma=0.5, C=0.5, alpha=1.0, beta=1.0, C2=100):
    
    if mins is None:
        mins = np.zeros(d)
    if maxs is None:
        maxs = np.ones(d)
    
    tau = K/np.sqrt(2*d)
    tau0 = K/np.sqrt(2*np.sqrt(d))
    
    best_objective_value = np.Inf
    best_chromosome = np.zeros((1, d))
    hist_of_best_objectives = np.zeros(number_of_iterations)
    last_best = np.zeros((1, d))
    
    current_population = np.zeros((population_size, d), dtype=np.float64)
    current_sigmas = np.zeros((population_size, d), dtype=np.float64)
    objective_values = np.zeros(population_size)

    
# set starting random population and sigmas
    current_population = np.random.uniform(low=mins, high=maxs, size=(d, population_size)).T
#     for i in range(population_size):
#         print(i)
#         while not conditions(current_population[i, :]):
#             current_population[i, :] = np.random.uniform(low=mins, high=maxs, size=d)
    
    current_sigmas = np.ones((population_size, d)) * sigma

# evaluating the objective function on the current population
#     for i in range(population_size):
#         objective_values[i] = objective_function(current_population[i, :]) + (C * 1.0)**alpha * conditions(current_population[i, :], beta)
    
# ---------------------
# main loop    
    for t in range(number_of_iterations):
        for i in range(population_size):
            objective_values[i] = objective_function(current_population[i, :]) + (C * (t+1.0))**alpha * conditions(current_population[i, :], beta, C2)
# parent selection
        children_population, children_sigmas = parent_selection(current_population, current_sigmas, 
                                                                objective_values, population_size, children_size)
        
# mutating children
        for i in range(children_size):
            epsilon0 = np.random.normal(loc=0.0, scale=tau0)
            children_sigmas[i, :] = children_sigmas[i, :] * np.exp(np.random.normal(loc=0.0, scale=tau, size=d) + epsilon0)
            children_population[i, :] = children_population[i, :] + np.random.normal(loc=0.0, scale=children_sigmas[i])
        
        children_population = np.clip(children_population.T, mins, maxs).T
        
# evaluating the objective function on the children
        children_objective_values = np.zeros(children_size)
        for i in range(children_size):
            children_objective_values[i] = objective_function(children_population[i, :]) + (C * (t+1.0))**alpha * conditions(children_population[i, :], beta, C2)
#         for i in range(children_size):
#             if conditions(children_population[i, :]):
#                 children_objective_values[i] = objective_function(children_population[i, :])
#             else:
#                 children_objective_values[i] = np.inf
            
        
# replacing the current population by given replacement
        current_population, current_sigmas, objective_values = replacement(current_population, children_population, 
                    current_sigmas, children_sigmas, objective_values, children_objective_values, population_size)
        
# recording some statistics
        hist_of_best_objectives[t] = objective_values[0]
        last_best = current_population[0, :]
        if best_objective_value > objective_values[0]:
            best_objective_value = objective_values[0]
            best_chromosome = current_population[0, :]
    return best_chromosome, hist_of_best_objectives, last_best
```

```python
# x* =  (78,33,29.995,45,36.7758), f(x*) = -30665.539
def G4_fun(x):
    return 5.3578547*(x[2]**2)+0.8356891*x[0]*x[4]+37.293239*x[0]-40792.141
def G4_cond(x, beta, C2):
    u = 85.334407+0.0056858*x[1]*x[4]+0.0006262*x[0]*x[3]-0.0022053*x[2]*x[4]
    v = 80.51249+0.0071317*x[1]*x[4]+0.0029955*x[0]*x[1]+0.0021813*(x[2]**2)
    w = 9.300961+0.0047026*x[2]*x[4]+0.0012547*x[0]*x[2]+0.0019085*x[2]*x[3]
    g = np.zeros(6)
    g[0] = u - 92
    g[1] = -u
    g[2] = v - 110
    g[3] = -v + 90
    g[4] = w - 25
    g[5] = -w + 20
    g = g * (g > 0) + (g > 0) * C2
    return (g ** beta).sum()
G4_mins = np.array([78,33,27,27,27]).reshape(-1, 1)
G4_maxs = np.array([102,45,45,45,45]).reshape(-1, 1)
G4_n = 5
```

```python
#  x* =  (14.095,0.84296), f(x*) = -6961.81388
def G6_fun(x):
    return (x[0]-10)**3 + (x[1]-20)**3
def G6_cond(x, beta, C2):
    g = np.zeros(2)
    g[0] = -(x[0]-5)**2 - (x[1]-5)**2 + 100
    g[1] = (x[0]-6)**2 + (x[1]-5)**2 - 82.81
    g = g * (g > 0) + (g > 0) * C2
    return (g ** beta).sum()
G6_mins = np.array([13, 0]).reshape(-1,1)
G6_maxs = np.array([100, 100]).reshape(-1,1)
G6_n = 2
```

```python
# x*=(2.171996, 2.363683, 8.773926, 5.095984, 0.9906548, 1.430574,1.321644, 9.828726, 8.280092, 8.375927)
# f(x*) = 24.3062091
def G7_fun(x):
    return x[0]**2+x[1]**2+x[0]*x[1]-14*x[0]-16*x[1]+(x[2]-10)**2+\
    4*(x[3]-5)**2+(x[4]-3)**2+2*(x[5]-1)**2+5*x[6]**2+\
    7*(x[7]-11)**2+2*(x[8]-10)**2+(x[9]-7)**2+45
def G7_cond(x, beta, C2):
    y = np.zeros(8)
    y[0] = 4*x[0]+5*x[1]-3*x[6]+9*x[7]-105
    y[1] = 10*x[0]-8*x[1]-17*x[6]+2*x[7]
    y[2] = -8*x[0]+2*x[1]+5*x[8]-2*x[9]-12
    y[3] = 3*(x[0]-2)**2+4*(x[1]-3)**2+2*x[2]**2-7*x[3]-120
    y[4] = 5*x[0]**2+8*x[1]+(x[2]-6)**2-2*x[3]-40
    y[5] = 0.5*(x[0]-8)**2+2*(x[1]-4)**2+3*x[4]**2-x[5]-30
    y[6] = x[0]**2+2*(x[1]-2)**2-2*x[0]*x[1]+14*x[4]-6*x[5]
    y[7] = -3*x[0]+6*x[1]+12*(x[8]-8)**2-7*x[9]
    y = y * (y > 0) + (y > 0) * C2
    return (y ** beta).sum()
G7_n = 10
G7_mins = (np.ones(G7_n) * (-10)).reshape(-1,1)
G7_maxs = (np.ones(G7_n) * 10).reshape(-1,1)
```

```python
#  x∗ = (1.2279713, 4.2453733) , f (x∗) = 0.095825
def G8_fun(x):
    return -(np.sin(2.0*np.pi*x[0]) ** 3 * np.sin(2.0*np.pi*x[1])) / (x[0]**3 * (x[0] + x[1]))
def G8_cond(x, beta, C2):
    g = np.zeros(2)
    g[0] = x[0] ** 2 - x[1] + 1
    g[1] = 1 - x[0] + (x[1] - 4.0) ** 2
    g = g * (g > 0) + (g > 0) * C2
    return (g ** beta).sum()
G8_n = 2
G8_mins = np.array([0, 0]).reshape(-1,1)
G8_maxs = np.array([10, 10]).reshape(-1,1)
```
