```python
import numpy as np

from pymoo.algorithms.nsga2 import NSGA2
from pymoo.model.problem import Problem
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
from pymoo.factory import get_problem
```

# ZDT2

```python
class MyProblem(Problem):
    def __init__(self):
        super().__init__(n_var=30,
                         n_obj=2,
                         n_constr=0,
                         xl=np.zeros(30),
                         xu=np.ones(30))

    def _evaluate(self, X, out, *args, **kwargs):
        g = 1 + 9 * X[:, 1:].sum(axis=1) / (X.shape[1]-1)
        f1 = X[:, 0]
        f2 = g * (1 - (X[:, 0] / g) ** 2)

        out["F"] = np.column_stack([f1, f2])
        


problem = MyProblem()

algorithm = NSGA2(pop_size=100)

res = minimize(problem,
               algorithm,
               ("n_gen", 100),
               verbose=True,
               seed=1)

plot = Scatter()
plot.add(res.F, color="red")
plot.show()
```

# ZDT4

```python
class MyProblem(Problem):
    def __init__(self):
        super().__init__(n_var=10,
                         n_obj=2,
                         n_constr=0,
                         xl=np.concatenate((np.array([0]), np.zeros(9)-5)),
                         xu=np.concatenate((np.array([1]), np.zeros(9)+5)))

    def _evaluate(self, X, out, *args, **kwargs):
        g = 1 + 10 * (X.shape[1]-1) + np.sum(X[:, 1:]**2 - 10 * np.cos(4 * np.pi * X[:, 1:]), axis=1)
        f1 = X[:, 0]
        f2 = g * (1 - np.sqrt(X[:, 0] / g))

        out["F"] = np.column_stack([f1, f2])
        


problem = MyProblem()

algorithm = NSGA2(pop_size=1000)

res = minimize(problem,
               algorithm,
               ("n_gen", 100),
               verbose=True,
               seed=1)

plot = Scatter()
plot.add(res.F, color="red")
plot.show()
```

# ZDT6

```python
class MyProblem(Problem):
    def __init__(self):
        super().__init__(n_var=10,
                         n_obj=2,
                         n_constr=0,
                         xl=np.zeros(10),
                         xu=np.ones(10))

    def _evaluate(self, X, out, *args, **kwargs):
        g = 1 + 9 * np.sqrt(np.sqrt(X[:, 1:].sum(axis=1) / (X.shape[1]-1)))
        f1 = 1 - np.exp(-4*X[:, 0]) * np.sin(6*np.pi*X[:, 0])**6
        f2 = g * (1 - (f1 / g) ** 2)

        out["F"] = np.column_stack([f1, f2])
        


problem = MyProblem()

algorithm = NSGA2(pop_size=100)

res = minimize(problem,
               algorithm,
               ("n_gen", 100),
               verbose=True,
               seed=1)

plot = Scatter()
plot.add(res.F, color="red")
plot.show()
```
